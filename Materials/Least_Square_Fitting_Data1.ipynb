```python
import numpy as np
from datascience import *
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from matplotlib.colors import ListedColormap
import seaborn as sns
```

```python
Linear_Data1 = Table().read_table('Linear_Data1.csv')
Linear_Data1.show()
```

```python
def fit_line(tbl, label_x, label_y):
    # Your code may need more than 1 line below here.
    xdata, ydata = tbl.column(0), tbl.column(1)
    def mse_f(slope, intercept):
        fitted = slope * xdata +intercept
        return np.mean((ydata - fitted) ** 2)
    values = minimize(mse_f)
    slope = values[0]
    intercept = values[1]
    return make_array(slope, intercept)
```

```python
m = fit_line(Linear_Data1, 'x', 'y')[0]
b = fit_line(Linear_Data1, 'x', 'y')[1]
print(f'The splope is {round(m, 7)} and the intercept is {round(b, 7)}.')
```
